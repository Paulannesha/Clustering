```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    print(x)
    return x
```

# k_means clustering

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
```

```python
df1.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df1 = pd.DataFrame(scaler.fit_transform(df1),columns = ['Balance', 'Qual_miles', 'cc1_miles', 'cc2_miles', 'cc3_miles',
       'Bonus_miles', 'Bonus_trans', 'Flight_miles_12mo', 'Flight_trans_12',
       'Days_since_enroll', 'Award?'] )
scaled_df1
```

```python
wcss = []
for i in range(1, 20):
    
    kmeans = KMeans(n_clusters=i,random_state=0 )
    kmeans.fit(scaled_df1)
    
    wcss.append(kmeans.inertia_)
print(wcss)    
plt.plot(range(1, 20), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
clusters_new = KMeans(6, random_state=42)
clusters_new.fit(scaled_df1)
```

```python
from sklearn.cluster import KMeans
clusters_new2 = KMeans(7, random_state=42)
clusters_new2.fit(scaled_df1)
```

# DBSCAN clustering

```python
from sklearn.cluster import DBSCAN
```

```python
array1 = df.values
array1
```

```python
stscaler = StandardScaler()
X = stscaler.fit_transform(array1)
X
```

```python
from sklearn.neighbors import NearestNeighbors

nearest_neighbors = NearestNeighbors(n_neighbors=11)
neighbors = nearest_neighbors.fit(X)
distances, indices = neighbors.kneighbors(X)
print(distances[0])
print(indices[0])

distances = np.sort(distances[:,10], axis=0)

fig = plt.figure(figsize=(5, 5))
plt.plot(distances)
plt.xlabel("Points")
plt.ylabel("Distance")
plt.grid()
```

```python
dbscan = DBSCAN(eps=2.2, min_samples=3)
clusters = dbscan.fit_predict(X)
```

```python
dbscan2 = DBSCAN(eps=2.8, min_samples=3)
clusters2 = dbscan2.fit_predict(X)
```

```python
dbscan.labels_
```

```python
dbscan2.labels_
```

```python
set(clusters)
```

```python
set(clusters2)
```

```python
df['cluster_dbscan']=dbscan.labels_
df['cluster_dbscan2']=dbscan2.labels_
df
```
